<h4>15. Programming Assignment</h4>

Write a program that asks the user to enter a number.

If the number is:

greater than 0, print "Positive" <br>
less than 0, print "Negative" <br>
exactly 0, print "Zero"

Use if, elif, and else.

````python
# Ask the user to enter a number.
number = int(input("Enter a number: "))

# Check whether the number is positive,
# negative, or zero.
if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
````

The computer checks the number one step at a time:

1. Is it greater than zero?
2. If not, is it less than zero?
3. If neither is true, the only number left is zero.

This is a great example of using if, elif, and else together to make a decision.
