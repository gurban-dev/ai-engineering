```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()
 
GROQ_API_KEY = os.environ.get("GROQ_API_KEY")

print(f"GROQ_API_KEY: {GROQ_API_KEY}")

# Initialize the client pointing to Groq's API endpoint
client = OpenAI(
    api_key=GROQ_API_KEY,
    base_url="https://api.groq.com/openai/v1"
)

def request_to_groq(prompt: str):
    try:
        # Make a request using a model hosted on Groq
        response = client.chat.completions.create(
            model="llama-3.1-8b-instant", # Groq supports Llama 3, Mixtral, Gemma, etc.
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7,
        )
    
        # Print the response.
        print("Groq response:")
        print(response.choices[0].message.content[:200])
    
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
request_to_groq("Create me an advanced python pygame shooting game")
```

<h3>Four Iteration Methods</h3>

High quality prompts are usually improved over time rather than being perfect <br>
on the first try.

Iteration 1: A very simple prompt

Tell me about dogs.

Result: The AI chatbot gives a general description of dogs.

Iteration 2: Add more detail

Tell me about dogs that make good pets for families.

Result: The answer becomes more focused on family-friendly breeds.

Iteration 3: Specify the audience

Explain which dogs make good family pets using words that new dog owners can <br>
understand.

Result: The explanation becomes simpler and easier to read.

Iteration 4: Specify the format

Explain which dogs make good family pets for people of all ages.

Include:
- Three dog breeds
- One fun fact about each breed
- A short conclusion

Result: The response is organised, easier to read, and contains exactly the <br>
information requested.

comparison clearer.

The relationship between a ruler and his subjects is analogous to that of <br>
