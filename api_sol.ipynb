<h4>15. Programming Assignment</h4>

Pretend you are making a small sentiment analyzer.

Write a Python program that:

Asks the user to type one word. <br>
If the word is "happy", print "Positive sentiment!" <br>
If the word is "sad", print "Negative sentiment!" <br>
Otherwise print "Neutral sentiment."

```python
!pip install textblob
```

```python
# Ask the user for one word.
word = input("Enter one word: ")

# Check if the word is happy.
if word == "happy":
    print("Positive sentiment!")

# Check if the word is sad.
elif word == "sad":
    print("Negative sentiment!")

# Otherwise, the sentiment is neutral.
else:
    print("Neutral sentiment.")

# The program asks for one word.

# Then it checks:

# 'Is it happy?'

# If not, it asks:

# 'Is it sad?'

# If neither answer is true, it prints Neutral.
```
